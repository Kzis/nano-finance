```python
df = pd.read_csv('data_nano_finance.csv')
df.head()
```

```python
content_list = df.content.tolist()

print(content_list)
```

```python
text = "‡∏Å‡∏π‡πâ ‡∏Å‡∏π‡πâ‡πÄ‡∏á‡∏¥‡∏ô ‡πÄ‡∏á‡∏¥‡∏ô‡∏Å‡∏π‡πâ ‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏Å‡∏π‡πâ‡πÄ‡∏á‡∏¥‡∏ô ‡∏à‡πà‡∏≤‡∏¢ ‡∏à‡∏≥‡πÄ‡∏õ‡πá‡∏ô ‡∏Ç‡∏≠‡∏Å‡∏π‡πâ ‡∏Ç‡∏≠‡∏Å‡∏π‡πâ‡πÄ‡∏á‡∏¥‡∏ô ‡∏Ñ‡∏∑‡∏ô ‡∏à‡πà‡∏≤‡∏¢‡∏™‡∏¥‡πâ‡∏ô‡πÄ‡∏î‡∏∑‡∏≠‡∏ô ‡πÉ‡∏Ñ‡∏£‡∏õ‡∏•‡πà‡∏≠‡∏¢‡∏Å‡∏π‡πâ ‡∏≠‡∏¢‡∏≤‡∏Å‡∏Å‡∏π‡πâ ‡∏û‡∏£‡∏∏‡πà‡∏á‡∏ô‡∏µ‡πâ"
text_shingles =  [text[i:i+3] for i in range(len(text))][:-3]
print(text_shingles)
```

```python
for document in content_list:
    # shingle and discard the last 5 as these are just the last n<5 characters from the document
    document = document.replace('?','').replace('u','').replace('b','')
    shingles = [document[i:i+3] for i in range(len(document))][:-3]
    print(shingles)
    
    # Jaccard distance is the size of set intersection divided by the size of set union
    jaccard_distance = len(set(shingles) & set(text_shingles)) / len(set(shingles) | set(text_shingles))
    print(jaccard_distance)
    
    print("===============")
```

```python
from datasketch import MinHash, MinHashLSH

for document in content_list:

    shingles = [document[i:i+3] for i in range(len(document))][:-3]
    set1 = set(shingles)
    set2 = set(text_shingles)
    
    m1 = MinHash(num_perm=128)
    m2 = MinHash(num_perm=128)
    
    for d in set1:
        m1.update(d.encode('utf8'))
    for d in set2:
        m2.update(d.encode('utf8'))


    # Create LSH index
    lsh = MinHashLSH(threshold=0.001, num_perm=128)

    result = lsh.query(m1)
    print("Approximate neighbours with Jaccard similarity > 0.1", result)
    print("===============")
```
